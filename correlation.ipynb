```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
#Read data
data = pd.read_excel('data_balanced_month.xlsx')
data.head()
```

```python
from matplotlib import pyplot
from pandas import read_csv
import numpy
# Path = r"C:\pima-indians-diabetes.csv"
names = ['month', "Age","Amount_netCN", "TermOfPaymentDays", 'GDP by State', 'Unemployment rate', 'ZIP', "DefaultStatus", 'Feq_default_ZipCode']
data = data[['month', "Age","Amount_netCN", "TermOfPaymentDays", 'GDP by State', 'Unemployment rate', 'ZIP', "DefaultStatus", 'Feq_default_ZipCode']]
```

```python
#https://www.tutorialspoint.com/machine_learning_with_python/machine_learning_with_python_correlation_matrix_plot.htm
correlations = data.corr().round(decimals=2)
#correlations.style.background_gradient(cmap='coolwarm').set_precision(2)
fig = pyplot.figure(figsize=(8, 6))
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = numpy.arange(0,len(names),1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names, rotation='vertical')
ax.set_yticklabels(names)

for i in range(len(names)):
    for j in range(len(names)):
        ax.text(j, i, correlations.iloc[i, j], ha='center', va='center', color='r')
#ax.figure.colorbar(im, ax=ax, format='% .2f')
pyplot.show()
```
